```python
c=df['vote_average'].mean()
c
```

```python
m=df['vote_count'].quantile(0.9)
m
```

```python
q=df.copy().loc[df['vote_count']>=m]
q.shape
```

```python
def weighted_rating(f,m=m,C=c):
    v=f['vote_count']
    R=f['vote_average']
    return (v/(v+m)*R)+(m/(m+v)*C)
```

```python
print(type(weighted_rating(df)))
```

```python
q['score']=q.apply(weighted_rating,axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

df['overview']  = df['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df['overview'])
tfidf_matrix.shape
```

```python
#Import linear kernel
from sklearn.metrics.pairwise import linear_kernel

#It will calculate  the similarity with all 4803 movies's review
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
cosine_sim.shape
```

```python
#Similarity score
print(cosine_sim[0])
print(cosine_sim[1])
```

```python
indices=pd.Series(df.index,index=df['title']).drop_duplicates()
```

```python
def get_recom(title,cosine_sim=cosine_sim):
    idx = indices[title]
    #Get get the pairwise similarity score of all movies with that movie
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    #print(sim_scores)
    #Start the movie based the similarity score
    sim_scores = sorted(sim_scores,key=lambda x:x[1],reverse = True)
    print(sim_scores[:10])
    print("\n")
    
    #Get only 10 top matching movies
    sim_scores = sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]
    #print(movie_indices)
    
    
    return df['title'].iloc[movie_indices]
    
```

```python
print(get_recom('Avatar'))
```

```python
print(get_recom('The Dark Knight Rises'))
```
