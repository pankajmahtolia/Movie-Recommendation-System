```python
C=df2['vote_average'].mean()
```

```python
m=df2['vote_count'].quantile(0.9)
m
```

```python
q_movies=df2.copy().loc[df2['vote_count']>=m]
q_movies.shape
```

```python
q_movies.head(5)
```

```python
def weighted_rating(x,m=m,C=C):
    v=x['vote_count']
    R=x['vote_average']
    #calculating based on the IMDB formula
    return (v/(v+m)*R)+(m/(m+v)*C)
```

```python
q_movies['score']=q_movies.apply(weighted_rating,axis=1)
```

```python
q_movies.head(3)
```

```python
q_movies=q_movies.sort_values('score',ascending=False)
q_movies[['title','vote_count','vote_average','score']].head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

df['overview'] = df['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df['overview'])
tfidf_matrix.shape
```

```python
#Import linear kernel
from sklearn.metrics.pairwise import linear_kernel

#
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
indices=pd.Series(df.index,index=df2['title']).drop_duplicates()
```

```python
def get_recom(title,cosine_sim=cosine_sim):
    idx = indices[title]
    #Get the pairwise similarity score of all movies with that movie
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    #Start the movie based the similarity score
    sim_scores = sorted(sim_scores,key=lambda x:x[1],reverse = True)
    
    
    #Get
    sim_scores = sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]
    
    
    
    return df['title_x'].iloc[movie_indices]
```

```python
get_recom('The Dark Knight Rises')
```
